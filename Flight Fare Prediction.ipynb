### 5.3.Feature Extraction

```python
#1.Converting string to int for duration column
import pandas as pd

def convert_duration(text):
    text = str(text)
    hours = 0
    minutes = 0

    if 'h' in text:
        hours = int(text.split('h')[0])
        text = text.split('h')[1]

    if 'm' in text:
        minutes = int(text.split('m')[0])

    return hours * 60 + minutes


df['Duration_minutes'] = df['Duration'].apply(convert_duration)

df.drop('Duration',axis=1,inplace=True) #Dropping original Duration Column
```

```python
#2.Deling With Date_of_Journey column
#Here we have to convert single column to different columns
df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'])


#Extracting Important Feature columns 
df['Journey_date']=df['Date_of_Journey'].dt.day 
df['Journey_month']=df['Date_of_Journey'].dt.month
df['Journey_year']=df['Date_of_Journey'].dt.year

#Dropping Date_of_Journey Column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
#3.Handling Dep_time column to extract features
df['Dep_Time']=pd.to_datetime(df['Dep_Time'],format='%H:%M')
df['Dep_Time'].dtype
df['dep_Hours']=df['Dep_Time'].dt.hour
df['dep_minutes']=df['Dep_Time'].dt.minute
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
#4.Handling Arrival_Time column to Extract Features
# Keep only HH:MM part
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]

# Convert to datetime format
df['Arrival_Time'] = pd.to_datetime(df['Arrival_Time'],format='%H:%M')

# Extract features
df['Arrival_Hour'] = df['Arrival_Time'].dt.hour
df['Arrival_Minute'] = df['Arrival_Time'].dt.minute

# Drop original column
df.drop('Arrival_Time', axis=1, inplace=True)
                                  
```

```python
#5.Handling Route column
#-Here i am dropping the route column because the route colum contains source point,Destination Point and Total Stops.So That Route is not needed.
df.drop('Route',axis=1,inplace=True)
```

### 5.4.Handling Outliers

```python
df.head()
```

```python
#Detecting Outliers in Duration_minutes Column by using boxplot
sns.boxplot(x='Duration_minutes',data=df)
```

```python
#STEP-1:Using Statistical Method finding outliers in Duration_minutes column
q1 = df['Duration_minutes'].quantile(0.25)
q3 = df['Duration_minutes'].quantile(0.75)

iqr = q3 - q1

lower_limit = q1 - 1.5 * iqr
upper_limit = q3 + 1.5 * iqr

print("Lower limit:", lower_limit)
print("Upper limit:", upper_limit)


#STEP-2:Techniques To Handle Outliers
#Here we are using Capping(Winsorization)-Replace Outliers with upper/lower limit
df['Duration_minutes']=df['Duration_minutes'].clip(lower_limit,upper_limit)


#Again Checking for outliers
sns.boxplot(x='Duration_minutes',data=df)
```

# 8.Model Comparision

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import pandas as pd
import numpy as np

models = {
    'Linear_Regression': LR,
    'Decision_tree': grid,
    'Random_Forest': RF,
    'Gradient_Boosting_Regressor': gbr,
    'XG_Boost': xgb,
    'Support_Vector_Regressor': svr,
    'CatBoost': cat,
    'Stacking': stack_model
}

results = []

for name, model in models.items():

    y_pred = model.predict(X_test)

    results.append({
        'Model': name,
        'Test_R2': r2_score(y_test, y_pred),
        'MAE': mean_absolute_error(y_test, y_pred),
        'RMSE': np.sqrt(mean_squared_error(y_test, y_pred)),
        'MSE': mean_squared_error(y_test, y_pred)
    })

results_df = pd.DataFrame(results)

results_df = results_df.sort_values(
    by='Test_R2',
    ascending=False
)

results_df = results_df.round(4)

pd.set_option('display.float_format', '{:.2f}'.format)

print(results_df)
```
